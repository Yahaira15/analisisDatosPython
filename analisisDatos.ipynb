#### 3. Generar graficos de las variables tipo 'categoricas'

Se genera una grafica con seaborn de las variables que sean de tipo categoricas

````python
# Se toman las variables categoricas y generamos los graficos

colCat = ['PeriodoAño', 'Deporte o Servicio', 'Nivel', 'Género',
          'Etnia', 'Tipo de Poblacion', 'Escenario', 'Instructor',
          'Nombre del Barrio', 'Municipio', 'Diversidad']

# Se transforman los datos del periodo y de fechas se pasa a años
data['PeriodoAño'] = pd.to_datetime(data['Periodo']).dt.year


fig, ax = plt.subplots(nrows=len(colCat), ncols=1, figsize=(10,55))
fig.subplots_adjust(hspace=2)


for i, col in enumerate(colCat):

#Mostrar las 10 categorias mas comunes
  top = data[col].value_counts().head(20)

  sns.barplot(x=top.index, y=top.values, ax=ax[i])
  ax[i].set_title(col)
  ax[i].set_xticklabels(ax[i].get_xticklabels(), rotation=75)
````

#### ***Analisis de las graficas***

Teniendo en cuenta que los datos que nos presentan son un reporte de las inscripciones echas en diferentes periodos en el Inde Envigado, estas inscripciones son a programas de actividad fisica y deportes
